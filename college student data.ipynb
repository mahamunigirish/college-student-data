```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('try me data sets 2.csv' ,usecols = ['Punctuality' , 'Lying'  ,'Internet usage' , 'Branded clothing','Entertainment spending','Spending on looks' ,'Spending on gadgets' ,'Spending on healthy eating' ,'Age' ,'Height' ,'Weight','Number of siblings' ,'Gender' ,'Village - town'  ])
```

```python
data.head()
```

```python
## Getting uniques of Lying coumn

sns.set_style('whitegrid')
sns.countplot(x='Lying',data=data)
```

```python
# Male lie more than female

sns.set_style('whitegrid')
sns.countplot(x='Lying',hue='Gender',data=data,palette='RdBu_r')
```

```python
# branded clothes wares females more than male

sns.set_style('whitegrid')
sns.countplot(x='Branded clothing',hue='Gender',data=data,palette='RdBu_r')
```

```python
# females more punctual than male
sns.set_style('whitegrid')
sns.countplot(x='Punctuality',hue='Gender',data=data,palette='RdBu_r')
```

```python
#females spends more times on looks
sns.set_style('whitegrid')
sns.countplot(x='Spending on looks',hue='Gender',data=data,palette='RdBu_r')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Internet usage',hue='Gender',data=data,palette='RdBu_r')
```

```python
# Males spends more times on gadgets
sns.set_style('whitegrid')
sns.countplot(x='Spending on gadgets',hue='Gender',data=data,palette='RdBu_r')
```

```python
# Which group of age spending more time on gadgets. - > 18-20
sns.jointplot(x='Age',y='Spending on gadgets',data=data,kind='hex')
```

```python
#IS Ages related to eating
sns.jointplot(x='Age',y='Spending on healthy eating',data=data,kind='hex')
```
